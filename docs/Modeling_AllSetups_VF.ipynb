```python
# SETUP 1
writefiletitle="../bifacialvf/data/Output/Setup1_VF.csv"
datestart = '2017-08-11 17:38:00'
dateend = '2017-08-24 13:59:00'
C = 0.82                      # GroundClearance(panel slope lengths). For tracking this is tilt = 0 hub height 
rtr = 1.5
# D = 0.51519                 # DistanceBetweenRows(panel slope lengths)

SRRLtoread = "..\\bifacialvf\data\SRRL_data.csv"
#SRRLtoread = "C:\Users\Silvana\Documents\GitHub\\bifacialvf\\bifacialvf\data\\SRRL_data.csv"
transFactor = 0         # TransmissionFactor(open area fraction)
cellRows = 609                # CellRows(# hor rows in panel)   <--> THIS ASSUMES LANDSCAPE ORIENTATION 
name = 'Golden Co'; lat=39.742; lng=-105.179;  tz = -7
beta = 10                   # PV tilt (deg)
sazm = 180                  # PV Azimuth(deg) or tracker axis direction
rowType = "interior"        # RowType(first interior last single)
PVfrontSurface = "glass"    # PVfrontSurface(glass or AR glass)
PVbackSurface = "glass"     # PVbackSurface(glass or AR glass)
albedo = 0.62               # ground albedo
tracking=False; backtrack=True
#Decimation routine Options== decimatestyle: Mean or Skip. decimatevlaue: 5,10,15,60
dataInterval = 1; decimate = True; decimatestyle = "Mean"; decimatevalue = 5 

simulateSRRL(SRRLtoread, writefiletitle,  beta, sazm, C, rtr= rtr, 
             rowType = rowType, transFactor = transFactor, cellRows = cellRows, 
             PVfrontSurface = PVfrontSurface, PVbackSurface = PVbackSurface,  
             albedo = albedo,  tracking = tracking, backtrack = backtrack, 
             decimate = decimate, decimatestyle = decimatestyle , decimatevalue = decimatevalue,      
             lat=lat, lng= lng, tz = tz, name = name, dataInterval = dataInterval,
             datestart = datestart, dateend = dateend)


#Load the results from the resultfile
from loadVFresults import loadVFresults
#print data.keys()
(data, metadata) = loadVFresults(writefiletitle)

data['IMT1']=data.loc[:,"No_126_RowBackGTI":"No_159_RowBackGTI"].mean(axis=1)
data['IMT2']=data.loc[:,"No_211_RowFrontGTI":"No_244_RowFrontGTI"].mean(axis=1)
data['IMT3']=data.loc[:,"No_296_RowBackGTI":"No_329_RowBackGTI"].mean(axis=1)
data['IMT4']=data.loc[:,"No_381_RowBackGTI":"No_414_RowBackGTI"].mean(axis=1)
data['IMT5']=data.loc[:,"No_466_RowFrontGTI":"No_499_RowFrontGTI"].mean(axis=1)
data['IMT6']=data.loc[:,"No_551_RowBackGTI":"No_584_RowBackGTI"].mean(axis=1)

data['GTIFrontavg'] = data[['IMT2', 'IMT4']].mean(axis=1)
data['GTIBackavg'] = data[['IMT1', 'IMT3', 'IMT4', 'IMT6']].mean(axis=1)

# Print the annual bifacial ratio.
frontIrrSum = data['GTIFrontavg'].sum()
backIrrSum = data['GTIBackavg'].sum()
print('The bifacial ratio for ground clearance {} and rtr spacing {} is: {:.1f}%'.format(C,rtr,backIrrSum/frontIrrSum*100))
#print("--- %s seconds ---" % (time.time() - start_time))
```
